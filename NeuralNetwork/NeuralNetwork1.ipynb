```python
# Generate some fake data with 3 features

from sklearn.datasets import make_classification

X, y = make_classification(n_features=3, n_redundant=0, n_informative=3,
                           random_state=42, n_classes=2, n_clusters_per_class=1)

y = y.reshape(-1, 1)

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# data preprocessing 
```

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
```

```python
#scale both training and testing 
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# one-hot encode the labels
from keras.utils import to_categorical 

y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
y_train_categorical
```

```python
# creating model 
```

```python
#defining model architecture
from keras.models import Sequential 

model = Sequential()
```

```python
# add first layer
from keras.layers import Dense
number_inputs = 3
number_hidden_nodes = 4
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
```

```python
#add final layers
number_classes = 2
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
# model summary 
model.summary()
```

```python
# compile the model
```

```python
#use categorical crossentropy for categorical data and mean squared errorfor regression 
# output layer here is using software for logistic regression (categorical)
# if your output layer activation was 'linear' then you may want to use ' mse' for loss
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# training the model
```

```python
# train the model using training data 
#training consists of updating our weights using our optimizer and loss function. 
# we choose 1000 iterations (loop) of training that called epochs.
# also choose to shuffle our training data and increase the detail printed out
  # during each training cycle
# fit the model
model.fit(X_train_scaled, 
         y_train_categorical,
         epochs=1000,
         shuffle=True,
         verbose=2)
```

```python
# Quantifying the model
```

```python
# use our testing data to validate our model. This is how we determine the validity of our model 
#(i.e. ability to predict new and previously unseen data points)
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f'Loss: {model_loss}, Accuracy: {model_accuracy}')
```

```python
# Making predictions with new data
```

```python
# we can use trained model to make predictions
import numpy as np
new_data = np.array([[0.2,0.3,0.4]])
print(f'Predicted class: {model.predict_classes(new_data)}')
```
