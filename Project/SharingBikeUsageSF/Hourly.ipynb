```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(os.path.join("data/GoBikeMerged2017_2018.csv"))
df.head()
```

```python
station = 'Embarcadero BART Station (Beale St at Market St)'
df2 = df.loc[(df['end_station_name'] == station) | (df['start_station_name'] == station)]
df2 = df2[['bike_id','end_station_name','start_station_name','start_time']]
df2
```

```python
times =  pd.DatetimeIndex(df2.start_time)
#groupby by hour and day
grouped_df = df2.groupby([times.hour,times.week,times.day,])
week_df = grouped_df.count()
week_df

week_df.to_csv('grouped_df_hour.csv')
```
